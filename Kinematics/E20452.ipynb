The QArm is a 4-DOF (Degree of Freedom) manipulator.

### Key Dimensions
- $L_1$: 0.14 m (Base height)
- $L_2$: 0.35 m (Shoulder to Elbow vertical offset)
- $L_3$: 0.05 m (Shoulder to Elbow horizontal offset)
- $L_4$: 0.25 m (Elbow to Wrist)
- $L_5$: 0.15 m (Wrist to End-effector)

### Simplified Parameters
- $λ_1$ = $L_1$
- $λ_2$ = $\sqrt{L_2^2​+L_3^2}$​
- $λ_3$ = $L_4 + L_5$
- $β$ = $tan^{−1}(L_3/L_2)$
\
- $\phi_1 = \theta_1$
- $\phi_2 = \theta_2 + \frac{\pi}{2} - \beta$
- $\phi_3 = \theta_3 + \beta$
- $\phi_4 = \theta_4$
